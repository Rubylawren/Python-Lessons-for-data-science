```python
# import the pandas library
import pandas as pd
```

```python
#read the datasets
df = pd.read_csv('big_mart_sales.csv')
```

```python
#view the first five rows
df.head(5)
```

**How to replace missing values** 

```python
# check for missing values
df.isna().sum()
```

```python
#get the rows with null values in the item weight
df.loc[df.Item_Weight.isna() == True]
```

```python
# fill int the null values in item weight with the mean
df.loc[(df.Item_Weight.isna() == True), 'Item_Weight'] = df.Item_Weight.mean()
```

```python
# fill the missing values in the outlet size with the mode
df.loc[(df.Outlet_Size.isna() == True), 'Outlet_Size'] = 'Median'
```

```python
#check for missing values again
df.isna().sum()
```

```python
df.head(5)
```

**UPDATING THE VALUES IN A COLUMN**

```python
#check the count of value in the item fat content column
df.Item_Fat_Content.value_counts()
```

```python
#create a mapping to update the item fat content column
mapping = {
    'Low Fat': 'LF',
    'Regular': 'R', 
    'LF': 'LF',
    'reg': 'R',
    'low fat': 'LF'
}
```

```python
#use the map function to updated the value
df.Item_Fat_Content = df.Item_Fat_Content.map(mapping)
```

```python
#count the new categories in the item fat content column
df.Item_Fat_Content.value_counts()
```

**HOW TO CREATE A COLUMN BY MODIFYING ANOTHER COLUMN**

```python
df.Item_MRP
```

```python
#create a new colum item_MRP_USD by dividing each value in the Item_MRP by 24 using apply function
df['Item_MRP_in_USD'] = df.Item_MRP.apply(lambda x: x/74)
```
