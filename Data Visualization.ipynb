```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("big_mart_sales.csv")
data
```

**LINE CHART**

```python
#mean price by item price
price_by_item = data.groupby("Item_Type").Item_MRP.mean()[:10]
```

```python
x = price_by_item.index.tolist()
y = price_by_item.values.tolist()

plt.figure(figsize=(14, 8))
plt.title('Mean price for each item type')

plt.xlabel('Item Type')
plt.ylabel('mean price')

plt.plot(x, y, marker='o')
```

**BAR CHART**

```python
#sales by outlet size
sales_by_outlet_size = data.groupby('Outlet_Size').Item_Outlet_Sales.mean()

#sort by sales
sales_by_outlet_size.sort_values(inplace=True)

x=sales_by_outlet_size.index.tolist()
y=sales_by_outlet_size.values.tolist()

#set axis labels
plt.xlabel("Outlet Size")
plt.ylabel("Sales")

#set title
plt.title("Mean sales for each outlet type")

plt.bar(x,y, color=['red', 'orange', 'magenta'])
```

**HISTOGRAM**

```python
#title
plt.title("Item MRP (price distribution) ")

#xlabel
plt.xlabel("Item MRP")

#ylabel
plt.ylabel("Frequency")

#plot the histogram
plt.hist(data['Item_MRP'], bins=20, color='lightblue')
```

**BOX PLOT**

```python
box_data = data['Item_Outlet_Sales']

#create outlier point shape
red_diamond = dict(markerfacecolor='r', marker='D')

#set title
plt.title('Item Sales Distribution')

#make the boxplot
plt.boxplot(box_data.values, labels=['Item Sales'], flierprops=red_diamond)
```

```python
box_data = data[['Item_Weight', 'Item_MRP']]

red_diamond = dict(markerfacecolor='r', marker='D')

#geberate subplots
fig, ax = plt.subplots()

#make the boxplot
plt.boxplot(box_data.values, labels=['Item Weight', 'Item MRP (price)'], flierprops=red_diamond)
```
