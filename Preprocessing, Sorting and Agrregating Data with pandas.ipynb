**SORTING WITH PANDAS**

```python
# import the pandas library
import pandas as pd
```

**Create a sample dataframe**

```python
#sample dataframe
df = pd.DataFrame({
    'roll_no': [102, 101, 104, 103, 105],
    'name': ['Ayotunde', 'Bobola', 'Daniel', 'Enoch', 'David'],
    'grade': ['B', 'B', 'A', 'C', 'A'],
    'marks': [15, 25, 20, 4, 22],
    'city': ['Ife', 'Ibadan', 'Lagos', 'Ekiti', 'Abia']
})
df
```

```python
#sort the data frame by roll_no column
df.sort_values(by= ['roll_no'])
```

```python
# sort the dataframe by grade in ascending order and marks in descending order
df.sort_values(by=['grade', 'marks'], ascending=[True, False], inplace=True)
```
